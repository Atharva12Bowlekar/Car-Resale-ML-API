#### Invalid Entry for Vehicle Age

```python
cleaned_car_df[cleaned_car_df["vehicle_age"]==0]
```

```python
cleaned_car_df["vehicle_age"] = cleaned_car_df["vehicle_age"].replace(0, 1)
```

#### Engineering features to give the model a better understanding of features

```python
cleaned_car_df["km_driven_per_year"] = round(cleaned_car_df["km_driven"] / cleaned_car_df["vehicle_age"], 0)
```

```python
cleaned_car_df["power_to_engine_ratio"] = round(cleaned_car_df["max_power"] * 1000 / cleaned_car_df["engine"], 2) 
# as engine figures are in cc and not litres
```

```python
cleaned_car_df.head()
```

#### Getting the skew values for all of the numerical columns

```python
numerical_columns = cleaned_car_df.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_columns
```

```python
num_col_to_skew = {}
```

```python
for num_col in numerical_columns:
    num_col_to_skew[num_col] = cleaned_car_df[num_col].skew().round(4)
```

```python
skew_df = pd.DataFrame.from_dict(num_col_to_skew, orient='index', columns=['Value'])
```

```python
skew_df
```

#### Normalising skewed columns

```python
log_cols = [
    "km_driven",
    "engine",
    "max_power",
    "km_driven_per_year",
    "selling_price_in_lakhs"
]

for col in log_cols:
    cleaned_car_df[col] = np.log1p(cleaned_car_df[col]).round(4)
```

#### Seats is a categorical feature

```python
cleaned_car_df["seats"] = cleaned_car_df["seats"].astype(str)
```

```python
cleaned_car_df.head()
```

#### One Hot Encoding all of the categorical columns

```python
categorical_columns = cleaned_car_df.select_dtypes(include=['object']).columns.tolist()
numerical_columns = cleaned_car_df.select_dtypes(exclude=['object']).columns.tolist()
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_columns)
    ],
    remainder="passthrough"
)
```

```python
categorical_columns
```

```python
numerical_columns
```

#### Training and testing models

```python
X = cleaned_car_df.drop("selling_price_in_lakhs", axis=1)
y = cleaned_car_df["selling_price_in_lakhs"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state = 21
)
```

```python
X_train.head()
```

```python
pipeline = Pipeline([
    ("preprocessing", preprocessor),
    ("model", XGBRegressor(
        colsample_bytree=0.8,
        gamma=0,
        learning_rate=0.05,
        max_depth=7,
        n_estimators=300,
        subsample=0.7
    ))
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
preds = pipeline.predict(X_test)
```

```python
preds[:5]
```

```python
model_path = Path().resolve() / "models" / "car_price_model.pkl"
```

```python
joblib.dump(pipeline, model_path)
```

```python
i = 0
count = 0

brand_count_exceed_dict = {}

for index, row in X_test.iterrows():
    brand = car_df[car_df.index == index]["brand"].iloc[0]
    if brand not in brand_count_exceed_dict:
        brand_count_exceed_dict[brand] = [0,0]
        
    brand_count_exceed_dict[brand][0] += 1
    change = abs(car_df[car_df.index == index]["selling_price"].iloc[0] - np.expm1(preds[i])*100000)
    percent_change = int(change) / int(np.expm1(preds[i])*100000)
    
    if percent_change < 0.2:
        pass
    else:
        count += 1
        brand_count_exceed_dict[brand][1] += 1
    i += 1
```

```python
round(count * 100 /len(X_test), 2)
```

```python
exceed_brand_dct = {}
for key, value in brand_count_exceed_dict.items():
    exceed_prcntge = round(value[1] * 100 / value[0], 2)
    exceed_brand_dct[exceed_prcntge] = key

dict(sorted(exceed_brand_dct.items()))
```
